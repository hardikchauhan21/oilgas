```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = '/Users/hardikchauhan/Downloads/oilgas.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 2: Check data types of the columns
print("Data types of columns in the dataset:")
print(data.dtypes)

# Step 3: Select only numerical columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Step 4: Handle missing values (optional: drop or fill)
numeric_data_cleaned = numeric_data.dropna()  # Drop rows with NaN values

# Step 5: Compute the correlation matrix
correlation_matrix = numeric_data_cleaned.corr()

# Step 6: Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
